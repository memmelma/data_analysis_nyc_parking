```python
import gmaps

new_york_coordinates = (40.75, -74.00)

apikey = "AIzaSyApj3xGPGx1naRs2DZiUlJ6moRftzWzTJU"
gmaps.configure(api_key=apikey)

file = 'unfinishedgeodata_07-09-2015_35674_Parking_Violations_Issued_-_Fiscal_Year_2016.csv'
datadir = '../data/nyc_parking_tickets/'
path = datadir + file
```

```python
#read data.csv
import csv

with open(path, 'r') as f:
    reader = csv.reader(f, quoting=csv.QUOTE_NONNUMERIC)
    data = list(reader)
#print(data)
```

```python
import pandas as pd

data = pd.read_csv(path, error_bad_lines=False)
data = data.convert_objects(convert_numeric=True).dropna()
data = data[:-1]#somehow last entry is not a location with lat and lan
print(len(data))
data = data.values.tolist()
#print(data)
```

```python
#heatmap
fig = gmaps.figure(center=new_york_coordinates, zoom_level=10)
fig.add_layer(gmaps.heatmap_layer(data))
fig
```

```python
#pointmap
fig = gmaps.figure(center=new_york_coordinates, zoom_level=10)
ticket_layer = gmaps.symbol_layer(data, fill_color="red", stroke_color="red", scale=1)
fig.add_layer(ticket_layer)
fig
```

```python
#directions
geneva = (46.2, 6.1)
montreux = (46.4, 6.9)
zurich = (47.4, 8.5)

fig = gmaps.figure()
geneva2zurich = gmaps.directions_layer(geneva, zurich, waypoints=[montreux])
fig.add_layer(geneva2zurich)
fig
```

```python
#folium map
import folium
from folium import features
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster

pubs_map = folium.Map(location=[40.742459, -73.971765], zoom_start=12)
HeatMap(data, radius = 20).add_to(pubs_map)
pubs_map
```
